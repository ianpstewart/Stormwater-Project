```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator, IndexFormatter 
  
import numpy as np

allData = pd.read_excel("Design Storm Hyeto.xlsx")
byYear = pd.DataFrame(allData)    
```

```python
ptm = {2:3.82, 10:5.53, 100:9.24}
```

```python
xticks = []
for i in range(241):
    if i%10 == 0:
        xticks.append(int(i/10))
    else:
        xticks.append('')

for year, depth in ptm.items():
    yearStr = str(year)
    byYear[yearStr] = depth*(byYear['p/pt2']-byYear['p/pt1'])
    title = str(year) + ' Year Design Storm Hyetograph'
    plt.figure();
    ax = byYear.plot.bar(x='t1', y=yearStr);
    plt.title(title);
    plt.xlabel('Time [hr]');
    ax.set_xticklabels(xticks)
    plt.ylabel("Rainfall Depth [in]");
    plt.savefig(title, dpi=300)
    
byYear.to_excel('output.xlsx')
```
