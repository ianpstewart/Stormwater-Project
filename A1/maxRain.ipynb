```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

allData = pd.read_csv("Rainfall History.csv")
byYear = pd.DataFrame(allData, columns = ['YEAR', 'IN'])    
```

```python
output = byYear.loc[byYear.groupby("YEAR")["IN"].idxmax()]
#print(output)
#output.to_excel('data.xlsx')
```

```python
plt.figure();
output.plot.scatter(x='YEAR', y='IN');
plt.title('Annual Maximum 24hr Rainfall Depths');
plt.xlabel('Year');
plt.ylabel("Rainfall Depth [in]");
plt.savefig('Annual Maximum Rainfall Depths.png', dpi=300)
```
